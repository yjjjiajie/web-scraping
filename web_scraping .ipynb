## Web Scraping 

From the website https://www.biopharmcatalyst.com/company/ABBV/news, I used web scraping to scrape the date of the news articles, as well as the text of the articles into a dataframe.

```python
import rpa as r
```

```python
r.init(turbo_mode = True)
```

```python
r.url('https://www.biopharmcatalyst.com/company/ABBV/news')
```

```python
links = [r.read(f'(//*[@class="card-divider"]/a)[{i}]/@href') for i in range(1, 6)]
links
```

```python
#Getting pages 2 to 60
r.url('https://www.biopharmcatalyst.com/company/ABBV/news?page=2')
links.extend([r.read(f'(//*[@class="card-divider"]/a)[{i}]/@href') for i in range(1, 6)])

while r.count('//*[@class="flex justify-between flex-1 sm:hidden"]/a/@href') == 2:
    r.url(r.read('(//*[@class="flex justify-between flex-1 sm:hidden"]/a/@href)[2]'))
    while not r.exist('(//*[@class="flex justify-between flex-1 sm:hidden"]/a/@href)'):
        continue
    links.extend([r.read(f'(//*[@class="card-divider"]/a)[{i}]/@href') for i in range(1, 6)])
```

```python
len(links)
```

```python
links
```

```python
date_time_list = []
contents_list = []

for link in links[:15]:
    r.url(link)
    while not r.exist("//*[@class='date']"): continue
    date_time_list.append(r.read("//*[@class='date']"))
    
    while not r.exist("//*[@class='article-content-body-only']/p//span"): continue
    contents_list.append(r.read("//*[@class='article-content-body-only']/p"))
```

```python
data = pd.DataFrame({"Date": date_time_list,
                    "Text": contents_list})


data["Company"] = "ABBV"
data = data[["Company","Date","Text"]]
data
```
